```python
import pandas as pd

# Read the data from the CSV file
df = pd.read_csv('com.csv')

# Create reversed rows
reversed_rows = []
for index, row in df.iterrows():
    reversed_row = row.copy()
    reversed_row["HomeTeam"], reversed_row["AwayTeam"] = row["AwayTeam"], row["HomeTeam"]
    reversed_row["FTHG"], reversed_row["FTAG"] = row["FTAG"], row["FTHG"]
    reversed_row["HTHG"], reversed_row["HTAG"] = row["HTAG"], row["HTHG"]
    reversed_row["HS"], reversed_row["AS"] = row["AS"], row["HS"]
    reversed_row["HST"], reversed_row["AST"] = row["AST"], row["HST"]
    reversed_row["HC"], reversed_row["AC"] = row["AC"], row["HC"]
    reversed_row["HF"], reversed_row["AF"] = row["AF"], row["HF"]
    reversed_row["HY"], reversed_row["AY"] = row["AY"], row["HY"]
    reversed_row["HR"], reversed_row["AR"] = row["AR"], row["HR"]
    reversed_rows.append(reversed_row)

# Convert list of reversed rows to DataFrame
df_reversed = pd.DataFrame(reversed_rows)

# Combine original and reversed DataFrames
df_combined = pd.concat([df, df_reversed]).sort_values(by=["HomeTeam", "AwayTeam"])

# Save to Excel
df_combined.to_excel("reordered_results.xlsx", index=False)

```

```python
import pandas as pd

# Read the CSV file
file_path = "com.csv"
df = pd.read_csv(file_path)

# Function to update FTR column with team names
def update_FTR(row):
    if row['FTR'] == 'H':
        return row['HomeTeam']
    elif row['FTR'] == 'A':
        return row['AwayTeam']
    else:
        return 'Draw'

# Apply the function to the dataframe
df['FTR'] = df.apply(update_FTR, axis=1)

# Save the updated dataframe to a new CSV file
output_path = "com_name.csv"
df.to_csv(output_path, index=False)

print(f"Updated CSV file saved as {output_path}")
```

```python
import pandas as pd

# Read the CSV file
file_path = "com.csv"
df = pd.read_csv(file_path, parse_dates=["Date"], dayfirst=True)

# Function to get the past 5 matches for a given team
def get_past_matches(team, date):
    past_matches = df[((df['HomeTeam'] == team) | (df['AwayTeam'] == team)) & (df['Date'] < date)]
    past_matches = past_matches.sort_values(by='Date', ascending=False).head(5)
    return past_matches

# Function to get data for a specific match and past matches for both teams
def get_match_data(date, home_team, away_team):
    match_date = pd.to_datetime(date, dayfirst=True)
    home_past_matches = get_past_matches(home_team, match_date)
    away_past_matches = get_past_matches(away_team, match_date)
    
    match_data = {
        "HomeTeamPastMatches": home_past_matches,
        "AwayTeamPastMatches": away_past_matches
    }
    
    # Give match_data a distinctive name
    match_key = f"match_data_{match_date.strftime('%Y%m%d')}_{home_team}_vs_{away_team}"
    return {match_key: match_data}

# Save the past matches to an Excel file
def save_to_excel(date, home_team, away_team, match_data_dict):
    match_date = pd.to_datetime(date, dayfirst=True)  # Convert date to datetime object
    formatted_date = match_date.strftime("%Y%m%d")
    file_name = f"{formatted_date}_{home_team}_vs_{away_team}.xlsx"
    match_data_key = list(match_data_dict.keys())[0]
    match_data = match_data_dict[match_data_key]
    
    with pd.ExcelWriter(file_name) as writer:
        match_data["HomeTeamPastMatches"].to_excel(writer, sheet_name=f"{home_team}_PastMatches", index=False)
        match_data["AwayTeamPastMatches"].to_excel(writer, sheet_name=f"{away_team}_PastMatches", index=False)
    print(f"Data saved to {file_name}")

# Example usage
match_date = "06/12/2014"
home_team = "Liverpool"
away_team = "Sunderland"

match_data_dict = get_match_data(match_date, home_team, away_team)
save_to_excel(match_date, home_team, away_team, match_data_dict)

# Get the key for the match data
match_data_key = list(match_data_dict.keys())[0]

print("Past 5 matches for Home Team:")
print(match_data_dict[match_data_key]["HomeTeamPastMatches"])

print("\nPast 5 matches for Away Team:")
print(match_data_dict[match_data_key]["AwayTeamPastMatches"])

```

```python
import pandas as pd

# Read the CSV file
file_path = "com.csv"
df = pd.read_csv(file_path, parse_dates=["Date"], dayfirst=True)

# Function to get the past 5 matches for a given team
def get_past_matches(team, date):
    past_matches = df[((df['HomeTeam'] == team) | (df['AwayTeam'] == team)) & (df['Date'] < date)]
    past_matches = past_matches.sort_values(by='Date', ascending=False).head(5)
    return past_matches

# Function to get past match data for all matches from 2008 onwards
def get_all_past_matches(df):
    all_matches_data = []

    for index, row in df[df['Date'] >= '2008-01-01'].iterrows():
        date = row['Date']
        home_team = row['HomeTeam']
        away_team = row['AwayTeam']

        home_past_matches = get_past_matches(home_team, date)
        away_past_matches = get_past_matches(away_team, date)
        
        match_data = {
            "Date": date,
            "HomeTeam": home_team,
            "AwayTeam": away_team,
            "HomeTeamPastMatches": home_past_matches,
            "AwayTeamPastMatches": away_past_matches
        }
        
        all_matches_data.append(match_data)

    return all_matches_data

# Save the past matches to an Excel file
def save_to_excel(all_matches_data):
    with pd.ExcelWriter("all_matches_data.xlsx") as writer:
        for match_data in all_matches_data:
            date_str = match_data["Date"].strftime("%Y%m%d")
            home_team = match_data["HomeTeam"]
            away_team = match_data["AwayTeam"]
            
            home_sheet_name = f"{date_str}_{home_team}_PastMatches"
            away_sheet_name = f"{date_str}_{away_team}_PastMatches"
            
            match_data["HomeTeamPastMatches"].to_excel(writer, sheet_name=home_sheet_name, index=False)
            match_data["AwayTeamPastMatches"].to_excel(writer, sheet_name=away_sheet_name, index=False)

    print("Data saved to all_matches_data.xlsx")

# Example usage
all_matches_data = get_all_past_matches(df)
save_to_excel(all_matches_data)
```
